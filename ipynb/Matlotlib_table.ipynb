### Tables with Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
   
val1 = ["{:X}".format(i) for i in range(10)] 
val2 = ["<{:X}>".format(i) for i in range(10)] 
val3 = [[c*r for c in range(10)] for r in range(10)] 
   
fig, ax = plt.subplots(figsize=(16, 6),facecolor='gray') 
ax.set_axis_off() 

table = ax.table( 
    cellText = val3,  
    rowLabels = val2,  
    colLabels = val1, 
    rowColours =["palegreen"] * 10,  
    colColours =["orange"] * 10,
    cellColours = [["yellow"] * 10] * 10,
    cellLoc ='center',  
#    loc ='upper left')         
    loc ='center')    
ax.set_title('matplotlib.axes.Axes.table() function Example', 
             fontweight ="bold") 

ax.text(0.45,0.07,'pippo')

# set_x/ylabel do not work because axis are off
ax.set_xlabel('x variable name')   
ax.set_ylabel('y variable name')
# set_facecolor does not work
ax.set_facecolor('red')

#plt.show() 
```

```python
fig, ax = plt.subplots() 
ax.set_axis_off() 

row_headers = val2
column_headers = val1
cell_text = val3

rcolors = plt.cm.BuPu(np.full(len(row_headers), 0.1))
ccolors = plt.cm.BuPu(np.full(len(column_headers), 0.1))

ax.set_facecolor('red')
#ax.set_backgroundcolor('blue')
table = ax.table( 
#the_table = plt.table(
                      cellText=cell_text,
                      rowLabels=row_headers,
                      rowColours=rcolors,
                      rowLoc='right',
                      colColours=ccolors,
                      colLabels=column_headers,
                      loc='center')
```
